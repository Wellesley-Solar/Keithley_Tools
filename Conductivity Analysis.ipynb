```python
import string, time, csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
%matplotlib notebook
```

```python
# V0=.035;
# Vinf=.054;
Iin=10E-11;

def simple_model(t_, C, V0, Vinf):
    Re=Vinf/Iin;
    Ri=Vinf*V0/(Iin*(Vinf-V0));
#     C=10^-7*5;
    return (Ri*Re)/(Ri+Re)*Iin*np.exp(-t_/((Re+Ri)*C))-Re*Iin*(np.exp(-t_/((Re+Ri)*C))-1);
```

```python
# Re
```

```python
temp=[]
with open('V3274p6s_darkon.csv', newline='') as csvfile:
    jvreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    # print(spamreader)
    for row in jvreader:
        temp.append(row)
```

```python
V_series=[float(i) for i in temp[0][1:]]
```

```python
t=3274.6;
t_series=np.linspace(1,len(V_series),len(V_series));
t_series=t_series*t/len(V_series);
```

```python
t_series
```

```python
# Simple Model Fitting
fitparam, fitparam_cov = optimize.curve_fit(simple_model, t_series, V_series,p0=[10E-8,0,0.06],bounds=([0,0,0.054],[1,0.035,1])) 
fitparam
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# ax.set_title('J-V Curve')
ax.plot(t_series, V_series, 'b:', label='data')
ax.plot(t_series, simple_model(t_series, *fitparam),'r--',label='fit')
# ax.axhline(y=0, color='k')
# ax.axvline(x=0, color='k')
ax.set_xlabel('Time (s)')
ax.set_ylabel('V_out (V)')   
ax.legend()
```
